# –î–æ–º–∞—à–Ω–µ–µ –∑–∞–¥–∞–Ω–∏–µ ‚Ññ1

```python
import numpy as np
from hashlib import sha256

def check(image, ref_hash):
    image_hash = sha256(np.ascontiguousarray(image)).hexdigest()
    if image_hash == ref_hash:
        print("All good \U0001F44D")
    else:
        raise RuntimeError("Image hash differ: " + image_hash)
```

## –ó–∞–¥–∞–Ω–∏–µ 1

–í—ã–≤–µ—Å—Ç–∏ –Ω–∞ —ç–∫—Ä–∞–Ω –≤—Ö–æ–¥–Ω–æ–µ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–µ, –∞ —Ç–∞–∫–∂–µ –æ—Ç–¥–µ–ª—å–Ω–æ –∫–∞–∂–¥—ã–π –∫–∞–Ω–∞–ª –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏—è

```python
import cv2
import skimage
import skimage.io
from matplotlib import pyplot as plt

src_image_url = 'https://drive.google.com/u/0/uc?id=1TVRYYk3KA4GeQropMHnfS3RrUSTK6fAe&export=download'
src_image = skimage.io.imread(src_image_url)

# Write your code here
src_image_r, src_image_g, src_image_b = cv2.split(src_image)

plt.figure(figsize=[6.4, 4.8*4])
plt.subplot(4,1,1)
plt.axis("off")
plt.title("Source image")
plt.imshow(src_image)

plt.subplot(4,1,2)
plt.axis("off")
plt.title("R-component")
plt.imshow(src_image_r, cmap='gray')

plt.subplot(4,1,3)
plt.axis("off")
plt.title("G-component")
plt.imshow(src_image_g, cmap='gray')

plt.subplot(4,1,4)
plt.axis("off")
plt.title("B-component")
plt.imshow(src_image_b, cmap='gray')

check(src_image, "db5c7a6ae4a0a6da43613c993c07997bf2f420cc8187aa3c1ffefca82a55e853")
check(src_image_r, "d3d0861f8b52481452bdbc25bfad3b64357b95346e9364bd4c6ec8fb83694507")
check(src_image_g, "9556656f92c4f738d02dee7214887647b63bac897a9f2979c830fb52bcf6c0e6")
check(src_image_b, "add87a423454669a49e86b9fc2a9794a95cdb723960aae6f6859636dc59b9483")
```

## –ó–∞–¥–∞–Ω–∏–µ 2
–ü—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞—Ç—å –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–µ –≤ –≥—Ä–∞–¥–∞—Ü–∏–∏ —Å–µ—Ä–æ–≥–æ –∏ –≤—ã–≤–µ—Å—Ç–∏ –Ω–∞ —ç–∫—Ä–∞–Ω

```python
import cv2

# Write your code here
gray = cv2.cvtColor(src_image, cv2.COLOR_RGB2GRAY)

plt.imshow(gray, cmap='gray')
plt.axis("off")

check(gray, "efd5a13b587c66e1db4bfae758a47414d6f59c1220299d27d0410b35536b12e8")
```

## –ó–∞–¥–∞–Ω–∏–µ 3
–ü—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞—Ç—å –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–µ –≤ –º–æ–¥–µ–ª—å YCrCb –∏ –≤—ã–≤–µ—Å—Ç–∏ –Ω–∞ —ç–∫—Ä–∞–Ω, –∞ —Ç–∞–∫–∂–µ –æ—Ç–¥–µ–ª—å–Ω–æ –≤—ã–≤–µ—Å—Ç–∏ –∫–æ–º–ø–æ–Ω–µ–Ω—Ç—É Y

```python
# Write your code here
ycrcb = cv2.cvtColor(src_image, cv2.COLOR_RGB2YCrCb)
ycrcb_y, _, _ = cv2.split(ycrcb)

plt.figure(figsize=[6.4, 4.8*2])
plt.subplot(2,1,1)
plt.axis("off")
plt.title("Source image")
plt.imshow(ycrcb)

plt.subplot(2,1,2)
plt.axis("off")
plt.title("Y-component")
plt.imshow(ycrcb_y, cmap='gray')

check(ycrcb, "42a7a4ca9be4c54f29ecf7b10a3f1060ac8d5d7c992cbbd908a1e3fff654d6a8")
check(ycrcb_y, "6dc61143b445519c63fcf57f9fd527665c5c7aad1bee35e72d17afdb9f12b0de")
```

## –ó–∞–¥–∞–Ω–∏–µ 4

–°—Ä–∞–≤–Ω–∏—Ç–µ –ø–æ–ª—É—á–µ–Ω–Ω–æ–µ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–µ –≤ –≥—Ä–∞–¥–∞—Ü–∏—è—Ö —Å–µ—Ä–æ–≥–æ –∏ –∫–æ–º–ø–æ–Ω–µ–Ω—Ç—É Y –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏—è –≤ —Ñ–æ—Ä–º–∞—Ç–µ YCrCb —Å –ø–æ–≥—Ä–µ—à–Ω–æ—Å—Ç—å—é 1

**–ü–æ–¥—Å–∫–∞–∑–∫–∞:** –î–ª—è —Å—Ä–∞–≤–Ω–µ–Ω–∏—è —Å –ø–æ–≥—Ä–µ—à–Ω–æ—Å—Ç—å—é –º–æ–∂–Ω–æ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å —Ñ—É–Ω–∫—Ü–∏—é [np.allclose](https://numpy.org/doc/stable/reference/generated/numpy.allclose.html)

```python
import numpy as np

# Write your code here
tolerance = 1

if np.allclose(gray, ycrcb_y, atol=tolerance):
    print("Images are kinda equal")
else:
    print("Nope, not the same")
```
