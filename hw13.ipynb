## Задание 1

Найти или сделать изображение шахматной доски в каком-нибудь окружении.

```python
import cv2
import numpy as np
import skimage
import skimage.io
from matplotlib import pyplot as plt

src_image_url = 'https://drive.google.com/u/0/uc?id=1XHb35CaTuuOs2gmA3_Z70NlfT3qPmV28&export=download'
src_image = skimage.io.imread(src_image_url)
```

## Задание 2

Преобразовать в оттенки серого.

```python
gray_image = cv2.cvtColor(src_image, cv2.COLOR_RGB2GRAY)

plt.imshow(gray_image, cmap='gray')
plt.axis("off");
```

## Задание 3

Найти углы шаблона с помощью findChessboardCorners.

```python
ret, corners = cv2.findChessboardCorners(gray_image, (6, 3), None)
```

## Задание 4

Отобразить найденные углы на изображении доски с помощью drawChessboardCorners.

```python
image_with_corners = src_image.copy()
cv2.drawChessboardCorners(image_with_corners, (6, 3), corners, ret)
plt.imshow(image_with_corners)
plt.axis("off");
```
