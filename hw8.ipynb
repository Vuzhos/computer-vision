## Задание 1

Выбрать тестовое изображение в оттенках серого и преобразовать его в бинарное с помощью пороговой сегментации.

```python
import cv2
import numpy as np
import skimage
import skimage.io
from matplotlib import pyplot as plt

src_image_url = 'https://drive.google.com/u/0/uc?id=1Z2Um_Yfyc9p9o82UyOTHOVwWaaE20EnE&export=download'
src_image = skimage.io.imread(src_image_url)
gray_image = cv2.cvtColor(src_image, cv2.COLOR_RGB2GRAY)
thresholded_image =  cv2.threshold(gray_image, np.iinfo(np.uint8).min, np.iinfo(np.uint8).max, cv2.THRESH_BINARY | cv2.THRESH_OTSU)[1]
```

## Задание 2

Выполнить эрозию полученного изображения с помощью круглого примитива размера 5х5 (примитив можно получить функцией cv.getStructuringElement).

```python
kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (5, 5))
eroded_image = cv2.erode(thresholded_image, kernel)
```

```python
plt.figure(figsize=[6.4 * 2, 4.8])
plt.subplot(1, 2, 1)
plt.imshow(thresholded_image, cmap='gray')
plt.axis("off")
plt.title("Original thresholded image")

plt.subplot(1, 2, 2)
plt.imshow(eroded_image, cmap='gray')
plt.axis("off")
plt.title("Eroded image");
```

## Задание 3

Выполнить дилатацию изображения с таким же примитивом.

```python
dilated_image = cv2.dilate(thresholded_image, kernel)
```

```python
plt.figure(figsize=[6.4 * 2, 4.8])
plt.subplot(1, 2, 1)
plt.imshow(thresholded_image, cmap='gray')
plt.axis("off")
plt.title("Original thresholded image")

plt.subplot(1, 2, 2)
plt.imshow(dilated_image, cmap='gray')
plt.axis("off")
plt.title("Dilated image");
```

## Задание 4

Выполнить операции размыкания исходного бинарного изображения.

```python
opened_image = cv2.dilate(cv2.erode(thresholded_image, kernel), kernel)
```

```python
plt.figure(figsize=[6.4 * 2, 4.8])
plt.subplot(1, 2, 1)
plt.imshow(thresholded_image, cmap='gray')
plt.axis("off")
plt.title("Original thresholded image")

plt.subplot(1, 2, 2)
plt.imshow(opened_image, cmap='gray')
plt.axis("off")
plt.title("Opened image");
```

## Задание 5

Выполнить операцию замыкания исходного бинарного изображения.

```python
closed_image = cv2.erode(cv2.dilate(thresholded_image, kernel), kernel)
```

```python
plt.figure(figsize=[6.4 * 2, 4.8])
plt.subplot(1, 2, 1)
plt.imshow(thresholded_image, cmap='gray')
plt.axis("off")
plt.title("Original thresholded image")

plt.subplot(1, 2, 2)
plt.imshow(closed_image, cmap='gray')
plt.axis("off")
plt.title("Closed image");
```

## Задание 6

Получить изображение, на котором выделены все границы исходного бинарного изображения.

```python
borders = thresholded_image - eroded_image
```

```python
plt.figure(figsize=[6.4 * 2, 4.8])
plt.subplot(1, 2, 1)
plt.imshow(thresholded_image, cmap='gray')
plt.axis("off")
plt.title("Original thresholded image")

plt.subplot(1, 2, 2)
plt.imshow(borders, cmap='gray')
plt.axis("off")
plt.title("All borders");
```
